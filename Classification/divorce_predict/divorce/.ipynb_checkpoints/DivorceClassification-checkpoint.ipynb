# Linear Regression

```python
X = divorce.drop('Class', axis=1)
y = divorce['Class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=100)
```

```python
#Create the linear model
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
print(lr.intercept_)
```

```python
lr.coef_
```

Create Predictions

```python
predictions = lr.predict(X_test)
```

```python
#Residuals Plot
sns.distplot(y_test - predictions)
```

```python
from sklearn import metrics
```

```python
metrics.mean_absolute_error(y_test, predictions)
```

```python
metrics.mean_squared_error(y_test, predictions)
```

```python
np.sqrt(metrics.mean_squared_error(y_test, predictions))
```

# K-Means Classification

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train, y_train)
```

```python
divorce_pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,divorce_pred))
print(classification_report(y_test,divorce_pred))
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
logmodel.score(X_train, y_train)
```

```python
logpredict = logmodel.predict(X_test)
```

```python
print(classification_report(y_test, logpredict))
```

# Support Vector Machines

```python
from sklearn import svm
```

```python
clf = svm.SVC()
```

```python
clf.fit(X_train, y_train)
```

```python
sv_pred = clf.predict(X_test)
```

```python
print(confusion_matrix(y_test,sv_pred))
print('\n')
print(classification_report(y_test,sv_pred))
```

# Decision Tree

```python
from sklearn import tree
```

```python
div_tree = tree.DecisionTreeClassifier()
```

```python
div_tree.fit(X_train, y_train)
```

```python
divTree_pred = div_tree.predict(X_test)
```

```python
print(confusion_matrix(y_test,divTree_pred))
print('\n')
print(classification_report(y_test,divTree_pred))
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
div_forest = RandomForestClassifier()
div_forest.fit(X_train, y_train)
```

```python
divforest_pred = div_forest.predict(X_test)
```

```python
print(confusion_matrix(y_test,divforest_pred))
print('\n')
print(classification_report(y_test,divforest_pred))
```
