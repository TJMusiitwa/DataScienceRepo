Download the data

```
import pandas as pd

url = "https://raw.githubusercontent.com/madmashup/targeted-marketing-predictive-engine/master/banking.csv"
marcamp = pd.read_csv(url)
```

```
import numpy as np
from sklearn import preprocessing
import matplotlib.pyplot as plt 
plt.rc("font", size=14)
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import train_test_split
import seaborn as sns
sns.set(style="white")
sns.set(style="whitegrid", color_codes=True)
```

```
marcamp['education'].unique()
```

```
marcamp['education']=np.where(marcamp['education'] =='basic.9y', 'Basic', marcamp['education'])
marcamp['education']=np.where(marcamp['education'] =='basic.6y', 'Basic', marcamp['education'])
marcamp['education']=np.where(marcamp['education'] =='basic.4y', 'Basic', marcamp['education'])
```

#Data Exploration

```
marcamp['y'].value_counts()
```

```
sns.countplot(x = 'y', data = marcamp)
plt.show()
```

```
count_no_sub = len(marcamp[marcamp['y']==0])
count_sub = len(marcamp[marcamp['y']==1])
pct_of_no_sub = count_no_sub/(count_no_sub+count_sub)
print("percentage of no subscription is", pct_of_no_sub*100)
pct_of_sub = count_sub/(count_no_sub+count_sub)
print("percentage of subscription", pct_of_sub*100)
```

#Data Visualisation

```
%matplotlib inline
pd.crosstab(marcamp.job,marcamp.y).plot(kind='bar')
plt.title('Purchase Frequency for Job Title')
plt.xlabel('Job')
plt.ylabel('Frequency of Purchase')
plt.savefig('purchase_fre_job')
```
