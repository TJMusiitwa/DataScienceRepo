```
 #Index Levels
  outside = ['G1', 'G1', 'G1', 'G2', 'G2', 'G2']
  inside = [1,2,3,1,2,3]
  hier_index = list(zip(outside, inside))
  hier_index = pd.MultiIndex.from_tuples(hier_index)
  
```

#Dealing with Missing Data

```
d = {'A':[12,24, np.nan], 'B':[45,90,0.98], 'C':[800, np.nan, np.nan]}
missdf = pd.DataFrame(d)
missdf
```

```
missdf.dropna(axis=0, thresh=2)
```

```
missdf.fillna(value=missdf.mean(), inplace=True)
```

```
missdf
```

#GroupBy Pandas

```
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
grpdata = pd.DataFrame(data)
grpdata
```

```
grpdata.groupby(by='Company').mean()
```

```
grpdata.groupby(by='Company').describe().transpose()
```
