```python
#Import the necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#Data Exploration

```python
data = pd.read_csv("sample_data/california_housing_train.csv")
```

```python
data.head()
```

```python
data.describe()
```

```python
data['total_bedrooms'].value_counts().plot(kind ='bar')
```

We can be able to show where the houses are distributed and located given the latitude and longitude values.

```python
plt.figure(figsize=(12,12))
sns.jointplot(x=data.latitude, y = data.longitude, size = 10)
plt.ylabel('Longitude')
plt.xlabel('Latitude')
plt.show()
sns.despine
```

Understanding how common factors are affecting house prices

```python
plt.scatter(data.median_house_value, data.population)
plt.title('Price Vs Population of the Area')
```

```python
plt.scatter(data.median_house_value, data.households)
plt.title('Price Vs Number of households')
```
