# Basic example

```python
import numpy as np
x = np.linspace(0,5,11)
y= x **2
```

```python
#Functional
plt.plot(x,y)
```

```python
year = [1,1.5,2,2.5,3,3.5]
gpa = [ 3.34, 3.22,3.27,3.52,3.38,3.10]
cgpa = [3.34,3.28,3.28,3.34,3.35,3.32]
```

```python
plt.subplot(1,2,1)
plt.plot(year, gpa)
plt.xlabel('Semester')
plt.ylabel('GPA')
plt.title('GPA against Semester')

plt.subplot(1,2,2)
plt.plot(year,cgpa,'r')
plt.xlabel('Semester')
plt.ylabel('CGPA')
plt.title('CGPA against Semester')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'g')
plt.subplot(1,2,2)
plt.plot(y,x,'r')
```

```python
#Object Oriented Method
figGPA = plt.figure()
axes = figGPA.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(year, gpa)
axes.set_title('Semester bottom vs GPA left')
```

```python
combined = plt.figure(figsize=(8,3))
axes1 = combined.add_axes([0.2, 0.2, 0.9, 0.9])
axes2 = combined.add_axes([0.2, 0.2, 0.9, 0.9])
axes1.plot(year, gpa, label='GPA', lw = 3, ls = '-.', marker = '+')
axes1.set_ylim([0,4])
axes2.plot(year, cgpa, label= 'CGPA', lw = 2, marker='*', ms=15)
axes2.set_ylim([0,4])
combined.legend(loc = 'center right', fontsize = 'large')

#To save the figure
#combined.savefig('results.png')
```
