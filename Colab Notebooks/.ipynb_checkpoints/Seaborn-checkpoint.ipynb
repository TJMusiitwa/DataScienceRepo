```python
#pip install seaborn
import seaborn as sns
%matplotlib inline
```

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head(3)
flights = sns.load_dataset('flights')
flights.head(3)
```

# Distribution Plots

```python
#Dist Plot
sns.distplot(tips['total_bill'])
```

```python
#JointPlot
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
#Pair Plot
sns.pairplot(tips, hue='sex')
```

```python
#rug plots
sns.rugplot(tips['total_bill'])
```

```python
sns.kdeplot(tips['total_bill'])
```

# Categorical Plots

```python
#BarPlot
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
#count plot
sns.countplot(x='sex', data=tips)
```

```python
#box plot
sns.boxplot(x='day', y='total_bill', data = tips, hue='sex', palette='seismic')
```

```python
#violinplot
sns.violinplot(x='day', y='total_bill', data=tips,hue='sex', split=True)
```

```python
#Strip plot
sns.stripplot(x='day', y='total_bill', data=tips, jitter=True, hue='sex', split=True)
```

```python
#Swarmplot
sns.swarmplot(x='day', y='total_bill', data = tips, hue='sex')
```

```python
#factor plot
sns.factorplot(x='day', y='total_bill', data=tips, kind='bar')
```

# Matrix Plots

```python
tc=tips.corr()
```

```python
sns.heatmap(tc, annot= True)
```

```python
fpvt = flights.pivot_table(index='month', columns='year', values='passengers')
fpvt
```

```python
sns.heatmap(fpvt, cmap='magma', linewidths=3)
```

```python
#cluster map
sns.clustermap(fpvt, cmap='coolwarm')
```

```python
sns.clustermap(tc)
```

# Grids

```python
iris = sns.load_dataset('iris')
```

```python
ip = sns.PairGrid(iris)
ip.map_diag(sns.distplot)
ip.map_lower(sns.kdeplot)
```

```python
sns.pairplot(iris)
```

```python
#Facetgrid
fg = sns.FacetGrid(data = tips, col='time', row='smoker')
fg.map(sns.distplot, 'total_bill')
```

# Regression Plots

```python
#Linear Plots
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex', row='time')
```

```python
sns.countplot(x='sex', data=tips)
sns.set_style('whitegrid')
sns.set_context('paper')
```
